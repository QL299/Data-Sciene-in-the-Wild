# Case 1

```python
#a
plt.scatter(individual['actual_average_screen_time'], individual['perceived_average_screen_time'],alpha =0.3)
plt.xlabel('Actual average screen time')
plt.ylabel('Perceived average screen time')
plt.savefig('1.png')
```

```python
#b

#remove negative values
data = individual[(individual['actual_average_screen_time']>0)]
#print(data.head())
#print(data.actual_average_screen_time)


#plot histogram to illustrate distribution
plt.hist(data['actual_average_screen_time'], bins = 'auto', color = 'purple', alpha = 0.7, rwidth = 0.85)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Actual average screen time')
plt.savefig('2.png')
```

```python
#C

original = individual['actual_average_screen_time'].to_list()
#1.Filling with mean
mean = np.mean(data['actual_average_screen_time'])
print(mean)
df_mean = individual['actual_average_screen_time'].to_list()
for n, i in enumerate(df_mean):
    if i <= 0:
        df_mean[n] = mean
print(df_mean)

#2.Filling with median
df_median = individual['actual_average_screen_time'].to_list()
median = np.median(data['actual_average_screen_time'])
print(median)
for n, i in enumerate(df_median):
    if i <= 0:
        df_median[n] = median
print(df_median)

#3. Filling with random value between 2 and 6
#print(df)
df_random = individual['actual_average_screen_time'].to_list()
for n, i in enumerate(df_random):
    if i <= 0:
        df_random[n] = random.randint(2.0,6.0)
print(df_random)

bins = np.linspace(-3, 15, 40)
plt.hist(original, bins, color = 'red', alpha=0.7, label='original')
plt.hist(df_mean, bins, color = 'skyblue', alpha=0.7, label='mean')
plt.hist(df_median, bins, color = 'green', alpha=0.7, label='median')
plt.hist(df_random, bins, color ='pink', alpha=0.7, label='random')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Actual average screen time')
plt.legend(loc='upper right')
plt.savefig('3.png')
```

```python
#d
re = np.random.normal(3.75, 1.25, 60)

#test with mean
#df = pd.DataFrame(individual['actual_average_screen_time'])
#print(df_1)
t1,p1 = stats.ttest_ind(df_mean,re)
print("t1 = " + str(t1))
print("p1 = " + str(p1))

#test with median
#print(df_2)
t2,p2 = stats.ttest_ind(df_median,re)
print("t2 = " + str(t2))
print("p2 = " + str(p2))


#test with random
#print(df_3)
t3,p3 = stats.ttest_ind(df_random,re)
print("t3 = " + str(t3))
print("p3 = " + str(p3))
```

```python
#b
cleaned_time = data['actual_average_screen_time']
#print(mean)
std = np.std(data['actual_average_screen_time'])
#print(std)
result = np.sum(cleaned_time > mean + std)
print(result)
```

```python
#c
perceived = individual['perceived_average_screen_time'].to_list()
actual = individual['actual_average_screen_time'].to_list()
bi_A = np.array(np.array(perceived) < 0, dtype=np.float)
bi_B = np.array(np.array(actual) > mean + std, dtype=np.float)
#print(bi_A)
#print(bi_B)
tab = pd.crosstab(bi_A, bi_B)
#print(tab)
p = stats.chi2_contingency(tab)
print("Chi square = " + str(p[0]))
print("p-value = " + str(p[1]))
```
